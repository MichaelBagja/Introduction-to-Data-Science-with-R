Penggunaan Vector dalam Rr
Vector merupakan struktur data yang menyimpan lebih dari 1 data di dalam R

```r
#penggunaan perintah Vector ( c )
print(c(1:5))

#perintah print digunakan untuk memunculkan nilai dari vector (c(1:5))
#perintah c digunakan untuk memanggil data vector
#perintah c(1:5) berfungsi untuk memanggil data vector antara angka 1 sampai 5
```

Perintah Summary dalam R

```r
#penggunaan perintah summary dapat memunculkan keseimpulan dalam data
#perintah summary memunculkan nilai dari nilai minimum, mean, median, modus, dan nilai maksimum dari data
#contoh penggunaan perintah summary

summary(c(1:5))

#perintah summary digunakan untuk memunculkan keseimpulan data
#perintah c(1:5) berfungsi untuk memanggil data vector dari angka 1 sampai 5
```
